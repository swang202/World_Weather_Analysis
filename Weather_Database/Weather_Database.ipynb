```python
import pandas as pd
import requests
import random
import numpy as np
from pprint import pprint
```

```python
#Create a new set of 2,000 random latitudes and longitudes.
lat = np.random.uniform(low=-90,high=90,size=2000)
lng = np.random.uniform(low=-180,high=180,size=2000)
lat_lngs = zip(lat,lng)
lat_lngs
```

```python
# Add the latitudes and longitudes to a list.
coordinates = list(lat_lngs)
coordinates
```

```python
#Get the nearest city using the citipy module.
from citipy import citipy

cities=[]

for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count.
len(cities)
```

```python
#Perform an API call with the OpenWeatherMap.
from config import weather_api_key 

# Test if the url is working
url="http://api.openweathermap.org/data/2.5/weather?&units=metric&q="
city="jamestown"
city_url=url+city+"&appid="+weather_api_key

print(city_url)

# Overlook of the data
city_test_data=requests.get(city_url).json()
pprint(city_test_data)
```

```python
#Retrieve the following information from the API call.

# Print the beginning of the logging.
print("Beginning Data Retrieval     ")
print("-----------------------------")

# Creat empty data
city_weather=[]

for i, city in enumerate(cities):
    
    try:
        print(f"Retrieving {city} data...")
        city_url=url+city+"&appid="+weather_api_key
        city_data=requests.get(city_url).json()
        city_name=city_data["name"]
        city_country=city_data["sys"]["country"]
        city_lat=city_data["coord"]["lat"]
        city_lng=city_data["coord"]["lon"]
        city_max_temp=city_data["main"]["temp_max"]
        city_min_temp=city_data["main"]["temp_min"]
        city_humidity=city_data["main"]["humidity"]
        city_clouds=city_data["clouds"]["all"]
        city_wind=city_data["wind"]["speed"]
        city_description=city_data["weather"][0]["description"]

        city_weather.append({"City": city_name,
                             "Country": city_country,
                              "Lat": city_lat,
                              "Lng": city_lng,
                              "Max Temp": city_max_temp,
                              "Min Temp": city_min_temp,
                              "Humidity": city_humidity,
                              "Cloudiness": city_clouds,
                              "Wind Speed": city_wind,
                              "Description": city_description})
    except:
        print("City not found.")
        pass

# Indicate that Data Loading is complete.
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
```

```python
# Convert the array of dictionaries to a Pandas DataFrame.
city_weather_df = pd.DataFrame(city_weather)
city_weather_df.head(10)
```

```python
# Save the DataFrame as a CSV file.
city_weather_df.to_csv("WeatherPy_Database.csv", index_label="City_ID")
```
